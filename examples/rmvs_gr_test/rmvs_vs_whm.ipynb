```python
import numpy as np
import swiftestio as swio
import matplotlib.pyplot as plt
```

```python
inparfile  = "param.swifter.in"
rmvsfile = swio.read_swifter_param(inparfile)
rmvsfile['BIN_OUT'] = 'bin.swifter.rmvs.dat'
whmfile = swio.read_swifter_param(inparfile)
whmfile['BIN_OUT'] = 'bin.swifter.whm.dat'

rmvsdat = swio.swifter2xr(rmvsfile)
whmdat = swio.swifter2xr(whmfile)
```

```python
rmvsdat['r'] = np.sqrt(rmvsdat['px']**2 + rmvsdat['py']**2 + rmvsdat['pz']**2)
whmdat['r'] = np.sqrt(whmdat['px']**2 + whmdat['py']**2 + whmdat['pz']**2)
```

```python
dr = rmvsdat['r'] - whmdat['r']
dr.sel(id=slice(100,109)).plot.line(x="time")
```
