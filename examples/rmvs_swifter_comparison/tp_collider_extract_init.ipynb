```python
from numpy.random import default_rng
```

```python
nclones = 10
xv_dispersion_factor = 0.01 
rng = default_rng()
clone_xv = (rng.standard_normal((nclones, 6)) - 0.5) * xv_dispersion_factor + 1.0
```

```python
px = 1.0
py = 2.0
pz = 3.0
vx = 4.0
vy = 5.0
vz = 6.0
jangofett = np.array([px, py, pz, -vx, -vy, -vz])
```

```python
jangofett
```

```python
clones = jangofett * clone_xv
```

```python
clones.shape
```

```python
clonenames = range(100, 100 + nclones)
```

```python
clonedat = dict(zip(clonenames,clones))
```
