```python
import numpy as np
import swiftest
import matplotlib.pyplot as plt
```

```python
swiftersim = swiftest.Simulation(param_file="param.swifter.in", codename="Swifter")
swiftersim.bin2xr()
```

```python
swiftersim.ds.plot.scatter(x='px', y='py', hue='id', hue_style="discrete", marker='.')
```

```python
swiftestsim = swiftest.Simulation(param_file="param.swiftest.in")
swiftestsim.bin2xr()
```

```python
swiftestsim.ds.plot.scatter(x='px', y='py', hue='id', hue_style="discrete", marker='.')
```

```python
swiftdiff = swiftestsim.ds - swiftersim.ds
swiftdiff = swiftdiff.rename_vars({'time' : 'time (y)','px' : 'delta x','py' : 'delta y'})
swiftdiff
```

```python
swiftdiff.plot.scatter(x='delta x', y='delta y', hue='id', hue_style="discrete", marker='.')
```

```python
plt.savefig('rmvsdiff.png')
```
