```python
import numpy as np
import swiftestio as swio
import matplotlib.pyplot as plt
```

```python
inparfile = 'param.swifter.in'
paramgr = swio.read_swifter_param(inparfile)
swifterdat = swio.swifter2xr(paramgr)
```

```python
swifterdat.plot.scatter(x='px', y='py', hue='id', hue_style="discrete", marker='.')
```

```python
config_file_name = 'config.swiftest.in'
config = swio.read_swiftest_config(config_file_name)
swiftestdat = swio.swiftest2xr(config)
```

```python
swiftestdat.plot.scatter(x='px', y='py', hue='id', hue_style="discrete", marker='.')
```

```python
swiftdiff = swiftestdat - swifterdat
swiftdiff = swiftdiff.rename_vars({'time' : 'time (y)','px' : 'delta x','py' : 'delta y'})
swiftdiff
```

```python
swiftdiff.plot.scatter(x='delta x', y='delta y', hue='id', hue_style="discrete", marker='.')
```

```python
plt.savefig('rmvsdiff.png')
```
